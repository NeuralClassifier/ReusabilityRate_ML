```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Loading Dataset

```python
cls_df = pd.read_excel('DataSets/classes.xlsx')
pkg_df = pd.read_excel('DataSets/packages.xlsx')
cls_df.drop(['Unnamed: 0'], axis=1, inplace=True)
pkg_df.drop(['Unnamed: 0'], axis=1, inplace=True)
pkg_df
```

### Pearson Correlation

```python
corr = {'cls':{}, 'pkg':{}}
corr['cls']['pearson'] = cls_df.corr(method='pearson', min_periods=1)
corr['pkg']['pearson'] = pkg_df.corr(method='pearson', min_periods=1)

f, axes = plt.subplots(2, 1, figsize=(20,20))
axes[0].title.set_text('Class')
axes[1].title.set_text('Package')
sns.heatmap(data=corr['cls']['pearson'], annot=True, ax=axes[0])
sns.heatmap(data=corr['pkg']['pearson'], annot=True, ax=axes[1])
```

### Spearman Correlation

```python
corr['cls']['spearman'] = cls_df.corr(method='spearman', min_periods=1)
corr['pkg']['spearman'] = pkg_df.corr(method='spearman', min_periods=1)

f, axes = plt.subplots(2, 1, figsize=(18,18))
axes[0].title.set_text('Class')
axes[1].title.set_text('Package')
sns.heatmap(data=corr['cls']['spearman'], annot=True, ax=axes[0])
sns.heatmap(data=corr['pkg']['spearman'], annot=True, ax=axes[1])
```

### Kendall Correlation

```python
corr['cls']['kendall'] = cls_df.corr(method='kendall', min_periods=1)
corr['pkg']['kendall'] = pkg_df.corr(method='kendall', min_periods=1)

f, axes = plt.subplots(2, 1, figsize=(18,18))
axes[0].title.set_text('Class')
axes[1].title.set_text('Package')
sns.heatmap(data=corr['cls']['kendall'], annot=True, ax=axes[0])

sns.heatmap(data=corr['pkg']['kendall'], annot=True, ax=axes[1])
```
