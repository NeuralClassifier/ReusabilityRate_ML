```python
import pandas as pd
import numpy as np
```

```python
project = pd.read_excel('/content/ReusabilityRate_ML/DataSets/projects_information.xlsx')
project.drop(['Unnamed: 0'],axis =1, inplace = True)
project
```

```python
packages = pd.read_excel('/content/ReusabilityRate_ML/DataSets/packages.xlsx')
packages.drop(['Unnamed: 0'],axis =1, inplace = True)
packages
```

```python
classes = pd.read_excel('/content/ReusabilityRate_ML/DataSets/classes.xlsx')
classes.drop(['Unnamed: 0'],axis =1, inplace = True)
classes
```

```python
classes.iloc[:,-1].mean()
```

```python
NoOfClasses = []
NoOfPackages = []
projectList = []
avgReuseRate_class = []
avgReuseRate_package = []
#projInClasses = list(classes.iloc[:,0])
for i in range(len(project)):
  proj = project.iloc[i,0]
  cls = classes[classes['Project']==proj]
  pckg = packages[packages['Project']==proj]
  projectList.append(proj)
  NoOfClasses.append(project.iloc[i,-1])
  NoOfPackages.append(project.iloc[i,-2])
  avgReuseRate_class.append(cls.iloc[:,-1].mean())
  avgReuseRate_package.append(pckg.iloc[:,-1].mean())

projectList=pd.DataFrame(projectList,columns=['Project'])
NoOfClasses=pd.DataFrame(NoOfClasses,columns=['NumberOfClasses'])
avgReuseRate_class=pd.DataFrame(avgReuseRate_class,columns=['AvgReuseRateClass'])
NoOfPackages=pd.DataFrame(NoOfPackages,columns=['NumberOfPackages'])
avgReuseRate_package=pd.DataFrame(avgReuseRate_package,columns=['AvgReuseRatePackage'])

NoCLass_to_rate = pd.concat([projectList,NoOfClasses,avgReuseRate_class],axis=1)
NoPackage_to_rate = pd.concat([projectList,NoOfPackages,avgReuseRate_package],axis=1)

NoCLass_to_rate.dropna(axis=0,inplace=True)
NoCLass_to_rate.reset_index(drop=True, inplace=True)
NoPackage_to_rate.dropna(axis=0,inplace=True)
NoPackage_to_rate.reset_index(drop=True, inplace=True)
```

```python
NoCLass_to_rate.to_excel('/content/NoClass_to_rate.xlsx')
NoPackage_to_rate.to_excel('/content/NoPackage_to_rate.xlsx')
```
